 # Gradient Descent

 - Automate the process of optimizing w and b using gradient descent

```python
import math, copy
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
from lab_utils_uni import plt_house_x, plt_contour_wgrad, plt_divergence, plt_gradients
```

```python
x_train = np.array([1.0, 2.0]) # size (1000 sq ft)
y_train = np.array([300.0, 500.0]) # price (1000s of dollars)
```

```python
def compute_cost(x, y, w, b):
    '''
    This function outputs a number that describes how well the model fits the training data
    '''

    m = x.shape[0] # number of training samples: in this case, 2
    cost = 0

    for i in range(m):
        f_wb = w * x[i] + b
        cost = cost + (f_wb - y[i])**2
    total_cost = 1 / (2 * m) * cost

    return total_cost
```

```python
def compute_gradient(x, y, w, b):
    """
    Computes the gradient for linear regression 
    Args:
      x (ndarray (m,)): Data, m examples 
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters  
    Returns
      dj_dw (scalar): The gradient of the cost w.r.t. the parameters w
      dj_db (scalar): The gradient of the cost w.r.t. the parameter b 
    """

    m = x.shape[0] # number of training samples: in this case, 2
    dj_dw = 0
    dj_db = 0

    # summation
    for i in range(m):
        f_wb = w * x[i] + b
        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i = f_wb - y[i] # notice how the partial derivative w.r.t. b does not have the extra "* x[i]"
        dj_dw += dj_dw_i # summation
        dj_db += dj_db_i
    dj_dw = dj_dw / m # in the end, divide both by 1/m to find the derivatives
    dj_db = dj_db / m

    return dj_dw, dj_db
```

```python
plt_gradients(x_train, y_train, compute_cost, compute_gradient)
plt.show()
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost_function, gradient_function):
    """
    Performs gradient descent to fit w,b. Updates w,b by taking 
    num_iters gradient steps with learning rate alpha
    
    Args:
      x (ndarray (m,))  : Data, m examples, training set
      y (ndarray (m,))  : target values
      w_in,b_in (scalar): initial values of model parameters weights and biases, likely set randomly
      alpha (float):     Learning rate
      num_iters (int):   number of iterations to run gradient descent
      cost_function:     function to call to produce cost
      gradient_function: function to call to produce gradient
      
    Returns:
      w (scalar): Updated value of parameter after running gradient descent
      b (scalar): Updated value of parameter after running gradient descent
      J_history (List): History of cost values
      p_history (list): History of parameters [w,b]
    """

    w = copy.deepcopy(w_in) # avoid modifying global variable w_in
    j_history = []
    p_history = []
    b = b_in
    w = w_in

    for i in range(num_iters):
      # compute the gradient using compute_gradient
      dj_dw, dj_db = gradient_function(x, y, w, b)

      b = b - alpha * dj_db
      w = w - alpha * dj_dw

      if i < 100000: # prevent resource exhaustion
        j_history.append(cost_function(x, y, w, b)) # history of cost values
        p_history.append([w, b]) # history of weights and biases (same index as j_history)

      # print cost every 10 intervals
      if i% math.ceil(num_iters/10) == 0:
        print(f"Iteration {i:4}: Cost {j_history[-1]:0.2e} ",
          f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e}  ",
          f"w: {w: 0.3e}, b:{b: 0.5e}")

    return w, b, j_history, p_history
```

```python
w_init = 0
b_init = 0

iterations = 10000
tmp_alpha = 1.0e-2

w_final, b_final, j_hist, p_hist = gradient_descent(x_train, y_train, w_init, b_init, tmp_alpha, iterations, compute_cost, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```

```python
# Final predictions with our optimized weights and biases

print(f"1000 sqft house prediction {w_final*1.0 + b_final:0.1f} Thousand dollars")
print(f"1200 sqft house prediction {w_final*1.2 + b_final:0.1f} Thousand dollars")
print(f"2000 sqft house prediction {w_final*2.0 + b_final:0.1f} Thousand dollars")
```
